```python
from docling.document_converter import DocumentConverter
import regex as re
converter = DocumentConverter()

def get_title_from_markdown(md: str) -> str | None:
    # İlk seviye başlık satırını bul (# Title)
    match = re.search(r"^#\s+(.+)$", md, re.MULTILINE)
    if match:
        return match.group(1).strip()
    return None

result = converter.convert("2210.03629v3.pdf").document.export_to_markdown()
title = get_title_from_markdown(result)
match = re.search(r"Abstract(.*?)References", result, re.DOTALL | re.IGNORECASE)
#result = " ".join(doc.page_content for doc in docs)
if match:
    result = match.group(1).strip()
else:
    result = result.strip()
print(title)
```
